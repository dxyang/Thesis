```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from utils import *
from glob import glob
import os
%matplotlib inline

import cv2
import ctypes
```

```python
G_imgs = np.load("generated_imgs.npy")
reconImgs = np.load("recon_imgs.npy")
hiddenImgs = np.load("hidden_imgs.npy")
mask = np.load("center_mask.npy")

G_imgs = 255*(G_imgs + 1.0)/2.0
reconImgs = 255*(reconImgs + 1.0)/2.0
hiddenImgs = 255*(hiddenImgs + 1.0)/2.0
mask = 255*(1 - mask)

print G_imgs.shape
print reconImgs.shape
print hiddenImgs.shape
print mask.shape
```
